```python
import numpy as np
```

```python
class perceptron(object):
  
    def __init__ (self, n_layers=1, n_inputs=1, verbose=3):
        self.n_layers = n_layers
        self.n_inputs = n_inputs
        self.verbose = verbose
        
        if self.verbose > 0: print 'Shell_init: layers ',n_layers,' inputs ', n_inputs
        
        self.layers = []
        for i in range(self.n_layers):
            self.layers.append(self.layer(i+1, self.verbose))
        
        self.build()
            
            
            
    def build (self):
        'Initialize weight matrix throught all layers'
        if self.verbose > 0: print 'Shell_build'
        inp = self.n_inputs
        for i in self.layers:
            i.build(inp)
            inp = i.size
               
            
            
    def add_neuron(self, tier, ftype=('t',0.5)):
        'Add neuron to specific tier'
        if self.verbose > 0: print 'Shell_add_neuron'
        self.layers[tier-1].add_neuron(ftype)
        self.build()

        
        
    def set_neuron(self, pos, ftype):
        if isinstance(pos, int): pos = [pos]
        assert len(pos) < 3
        if len(pos) == 1:
            for i in self.layers[pos[0]-1].neurons:
                i.ftype = ftype
            if self.verbose > 0: print 'Changing neurons in: tier', pos[0], 'to function:', ftype
        else:
            self.layers[pos[0]-1].neurons[pos[1]-1].ftype = ftype
            if self.verbose > 0: print 'Changing neuron: tier', pos[0], '#' + str(pos[1]), 'to function:', ftype
    
    
    def show_weights(self):
        for i in self.layers:
            i.show_weights()
        
            
            
    def push(self, x):
        'Input value and propagate it thruogh the network'
        if type(x) is not np.ndarray:
            x = np.asarray(x)
            x = x.reshape(-1, self.n_inputs)
#         assert x.shape[1] == self.n_inputs
        
        out = np.zeros_like((x.shape[0], 1))  # need to implement n_outputs
        for n in range(x.shape[0]):
            if self.verbose > 0: print '-------------------\nShell_push \ninput vector:\n', x[n]
            y = x[n]
            for i in self.layers:
                y = i.push(y)
            out[n] = y
            if self.verbose > 0: print 'output vector:\n', y
        return out
            
    
    
    
    class layer(object):
        
        def __init__ (self, tier, verbose):
            
            self.size = 1
            self.tier = tier
            self.verbose = verbose
            
            if self.verbose > 1: print '\tLayer_init: tier',self.tier,'size ', self.size
            
            self.neurons = []
            for i in range(self.size):
                self.neurons.append(self.neuron(('t', 0.5),i+1, self.verbose))
            
            
        def build(self, inputs):
            if self.verbose > 1: print '\tLayer_build: ', 'tier', self.tier
            self.w = np.random.random_sample((inputs, self.size))
            
        def show_weights(self):    
            if self.verbose > 1: print '\tLayer ', self.number, ' w-matrix:\n\t', self.w
            
        def add_neuron(self, ftype):
            if self.verbose > 1: print '\tLayer_add_neuron'
            self.size += 1
            self.neurons.append(self.neuron(ftype, self.size, self.verbose))
            
            
        def push(self, x):
            if self.verbose > 1: print '\tLayer_push: tier',self.tier,'\n\twmatrix:\n', self.w
            
            x = x.dot(self.w)
            for i in enumerate(self.neurons):
                x[i[0]] = i[1].push(x[i[0]])
            
            if self.verbose > 1: print '\toutput vector\n\t', x
            
            return x
                
                
        
        
        class neuron(object):
            
#             @staticmethod
#             def threshold(x, a):
#                 print '   neuron_func_threshold'
#                 return 1 if x >= a else 0
            
#             @staticmethod
#             def sigmoid(x, a):
#                 print 'neuron_func_sigmoid'
#                 return 1/(1+np.exp(-x))
             
            @staticmethod
            def func(ftype, x):
                if ftype[0] == 't':
                    return 1 if x >= ftype[1] else 0
                elif ftype =='s':
                    return 1/(1+np.exp(-x))

            
            
            def __init__(self, ftype, num, verbose):
                self.ftype = ftype
                self.idx = num
                self.verbose = verbose
                if self.verbose > 2: print '\t\tNeuron_init: type', ftype
                
                
            def push(self, x):
                out = perceptron.layer.neuron.func(self.ftype, x)
                if self.verbose > 2: print '\t\tNeuron_push: #'+str(self.idx)+', type:',self.ftype,'\n\t\t', x, '->', out
                return out
            
```
